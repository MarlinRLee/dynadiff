```python
import os
import pandas as pd
import numpy as np

def process_and_concatenate_tsv_files(directory_path, ses_id = "01"):
    all_dfs = []

    # Iterate through each file in the directory
    for filename in os.listdir(directory_path):
        if filename.endswith(".tsv"):# and f"session{ses_id}" in filename:
            print(filename)
            file_path = os.path.join(directory_path, filename)
            try:
                # Read the TSV file into a pandas DataFrame
                df = pd.read_csv(file_path, sep='\t')

                # Remove rows where all values are zero
                # We check if all values in a row are equal to 0, then select rows where this is not true.
                df_filtered = df[~(df == 0).all(axis=1)]

                # Append the filtered DataFrame to our list
                all_dfs.append(df_filtered)
            except Exception as e:
                print(f"Error processing file '{filename}': {e}")

    # Concatenate all collected DataFrames into one if any were processed
    if all_dfs:
        final_df = pd.concat(all_dfs, ignore_index=True)
        # Convert the final DataFrame to a NumPy array
        final_array = final_df.to_numpy()
        print(f"Successfully processed {len(all_dfs)} TSV files.")
        print(f"Concatenated array shape: {final_array.shape}")
        return final_array
    else:
        print("No TSV files found or successfully processed in the specified directory.")
        return None

your_directory = "/scratch.global/lee02328/dynadiff/nsddata/nsddata_timeseries/ppdata/subj01/func1pt8mm/design/" # CHANGE THIS TO YOUR ACTUAL PATH

# Call the function to process your files
result_array = process_and_concatenate_tsv_files(your_directory, "03")

len(result_array)
```
